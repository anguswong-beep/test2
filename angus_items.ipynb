```python
price_types = [ 'reg','sale', 'tpr']
```

```python
sale_types = ['Sale','Refunded','voided']
```

```python
sale_types_weights = [.7,.1,.2]
```

```python
random.seed(42)

def generate_transactions(base_itemss,start_date,end_date,dtc=100,max_itemss=30,mqpi=10,minda=.25,maxda=1.0):
    '''
     Generate a random set of transactions using the base dataset of itemss

     Parameter:
        base_itemss(list): list of base itemss
        start_date(datetime): sd
        end_date(datetime): sd
        dtc(int): desired transaction count
        max_itemss(int): max num of itemss per transaction
        mgpi(int): max quantity of items
        minda(float): min discount amount of items
        maxda(float): max discount amount of items
    '''

    date_range = pd.date_range(start_date,end_date, freq='d')
    transaction_list = []
    for store in store_meta:
        last_transaction_id = 1
        for day in date_range:
            for tx_id in range(last_transaction_id,dtc + 1 + last_transaction_id):
                num_itemss = random.randint(1,max_itemss)
                for seq, _ in enumerate(range(num_itemss)):
                    items = random.choice(base_itemss)
                    qty = random.randint(1,mqpi)
                    price = items['price'] 
                    retail_price = price
                    price_type = random.choice(price_types)
                    if price_types != 'reg':
                        discount_amount = round(random.uniform(minda,maxda),2)
                        price = price - discount_amount
                    sales = round(qty * price,2)
                    sale_type = random.choices(sale_types, sale_types_weights,k=1)[0]

                    transaction_list.append({
                        'storeid':store['storeid'],
                        'store_number':store['store_number'],
                        'store_name':store['store_name'],
                        'sale_date':day.date(),
                        'transaction_id':tx_id,
                        'sale_type':sale_type,
                        'line_number':seq+1,
                        'product_code':items['product_code'],
                        'product_description':items['product_description'],
                        'price_type':price_type,
                        'qty':qty,
                        'retail_price': retail_price,
                        'price':price,
                        'sales':sales,
                        'cost':items['cost'],
                        'vendor_id': items['vendor_id'],
                        'vendor_name': items['vendor_name'],
                        'sub_department': items['sub_department'],
                        'sub_department_description': items['sub_department_description'],
                        'category': items['category'],
                        'category_description': items['category_description']

                    })
            last_transaction_id = dtc+1
    return pd.DataFrame(transaction_list)
    
```

```python
def load_dataset_to_postgres(ds, table_name, schema='public'):
    import psycopg2
    from io import StringIO

    conn = psycopg2.connect(f'dbname={database} user={username} password={password} host={host}')
    cur = conn.cursor()
    buffer = StringIO()
    ds.to_csv(buffer, index=False, header=False)
    buffer.seek(0)

    cur.copy_from(buffer, f'{table_name}', sep=',')
    conn.commit()
    cur.close()
    conn.close()
```

```python
load_dataset_to_postgres(df,'sales')
```
